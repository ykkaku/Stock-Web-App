```python
import csv
import pandas as pd
```

```python
companylist=list()
with open('500.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    for row in reader:
        companylist.append(row[0])
```

```python
companyfull=list()
with open('500.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    #next(reader)
    for row in reader:
        companyfull.append(row)
        
```
