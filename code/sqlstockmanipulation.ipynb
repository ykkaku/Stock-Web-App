```python
def checkmonthindustry(Industry,month,cursor):
    sqlcode=('Select * From CompanyInfo Where Sector="'+Industry+'"')
    cursor.execute(sqlcode)
    Industry = cursor.fetchall()
    allcomp=list()
    newdict=dict()
    for i in Industry:
        try:
            sql=('Select * From '+i[1]+' Where Date Like "'+month+'"')
            cursor.execute(sql)
            companymonth = cursor.fetchall()
            df = pd.DataFrame(companymonth)
            dffinal=df.rename(columns={0:'Date',1:'Open',2:'High',3:'Low',4:'Close',5:'Volume',6:'Dividends',7:'Stock Splits'})
            monthigh=max(dffinal['High'])
            monthlow=min(dffinal['Low'])
            volumelow=min((dffinal['Volume']))
            volumehigh=max((dffinal['Volume']))
            volumerange=str(volumelow)+'-'+str(volumehigh)
            totaldiv=sum(dffinal['Dividends'])
            symbols=i
            #names=['Symbols','MonthHigh','MonthLow','VolumeHigh','VolumeLow','VolumeRange','TotalDividends']
            svalues=[]
            svalues.extend([monthigh,monthlow,volumehigh,volumelow,volumerange,totaldiv])
            newdict[i[1]]=svalues
        except:
            continue
        
    df11 = pd.DataFrame(newdict).T
    df11=df11.rename(columns={'index':'Ticker',0:'MonthHigh',1:'MonthLow',2:'VolumeHigh',3:'VolumeLow',4:'VolumeRange',5:'TotalDividends'})
    return df11
```

```python
#return stock information of the particular industry in the selected month  
checkmonthindustry('Communication Services',month,cursor)
```
